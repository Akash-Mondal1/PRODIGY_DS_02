```python
top_cities = df['City'].value_counts().head(5)
plt.pie(top_cities, labels=top_cities.index, autopct='%1.2f%%')
plt.title('Top 5 Cities Distribution')
plt.show()
```

```python
# Calculate number of restaurants by country
restaurant_count_by_country = final_df.groupby('Country').size().reset_index(name='Restaurant Count').head(6)

# Calculate average rating by country
avg_rating_by_country = final_df.groupby('Country')['Aggregate rating'].mean().reset_index()

# Merge average rating and restaurant count dataframes
country_data = pd.merge(avg_rating_by_country, restaurant_count_by_country, on='Country')

# Visualize average aggregate rating and restaurant count by country
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='Country', y='Aggregate rating', data=country_data)
plt.title('Average Aggregate Rating by Country')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(x='Country', y='Restaurant Count', data=country_data)
plt.title('Number of Restaurants by Country')

plt.tight_layout()
plt.show()
```
